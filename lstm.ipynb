## scraping penjelasan tentang LSTM
source : https://colah.github.io/posts/2015-08-Understanding-LSTMs/index.html


```python
import requests
import string
import re
from bs4 import BeautifulSoup

# mengambil data dari source menggunakan requests
source = requests.get("https://colah.github.io/posts/2015-08-Understanding-LSTMs/index.html").text
isi = BeautifulSoup(source)

#membuang tag yang tidak penting (tidak memiliki informasi didalamnya dan menyisakan tag dengan informasi penting )
for i in isi(['style','li','ul','script']):
    i.decompose()

teks = "".join(isi.stripped_strings)
print(teks)
```

```python
teks = teks.lower()
teks = re.sub(r"\d+","",teks) #=> menghilangkan angka menggunakan re
teks = teks.translate(str.maketrans("","",string.punctuation))
teks = teks.strip() #=> menghapus karakter kosong

print(teks)
```

```python
with open('colah_lstm.txt','w') as filetxt:
    filetxt.write(teks)
```
